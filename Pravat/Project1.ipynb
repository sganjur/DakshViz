```python
import pandas as pd
mdf = pd.read_csv("./Data/voter-survey-ed.csv")
```

```python
def getCandidateVsParty(df):
    subdf = df[['The Candidate',"Candidate's Party"]]
    
    cand_imp_total = len(subdf.loc[subdf['The Candidate'] == 'Important'].index)
    cand_not_imp_total = len(subdf.loc[subdf['The Candidate'] == 'Not Important'].index)
    cand_very_imp_total = len(subdf.loc[subdf['The Candidate'] == 'Very Important'].index)
    party_imp_total = len(subdf.loc[subdf["Candidate's Party"] == 'Important'].index)
    party_not_imp_total = len(subdf.loc[subdf["Candidate's Party"] == 'Not Important'].index)
    party_very_imp_total = len(subdf.loc[subdf["Candidate's Party"] == 'Very Important'].index)
    
    cand_total = cand_imp_total + cand_not_imp_total+cand_very_imp_total  
    party_total = party_imp_total + party_not_imp_total + party_very_imp_total
    totals = {"Candidate":{
                           "Important":(cand_very_imp_total + cand_imp_total)/cand_total, 
                           "Not Important":cand_not_imp_total/cand_total,
                           },
             "Party":{
                      "Important":(party_imp_total + party_very_imp_total)/party_total, 
                      "Not Important":party_not_imp_total/party_total,
                           }}
    return pd.DataFrame.from_dict(totals)
```

```python
def getCandidateCasteVsParty(df):
    subdf = df[["Candidate's Caste/Religion","Candidate's Party"]]
    
    cand_imp_total = len(subdf.loc[subdf["Candidate's Caste/Religion"] == 'Important'].index)
    cand_not_imp_total = len(subdf.loc[subdf["Candidate's Caste/Religion"] == 'Not Important'].index)
    cand_very_imp_total = len(subdf.loc[subdf["Candidate's Caste/Religion"] == 'Very Important'].index)
    party_imp_total = len(subdf.loc[subdf["Candidate's Party"] == 'Important'].index)
    party_not_imp_total = len(subdf.loc[subdf["Candidate's Party"] == 'Not Important'].index)
    party_very_imp_total = len(subdf.loc[subdf["Candidate's Party"] == 'Very Important'].index)
    
    cand_total = cand_imp_total + cand_not_imp_total+cand_very_imp_total  
    party_total = party_imp_total + party_not_imp_total + party_very_imp_total
    totals = {"Caste/Religion":{
                           "Important":(cand_very_imp_total + cand_imp_total)/cand_total, 
                           "Not Important":cand_not_imp_total/cand_total,
                           },
             "Party":{
                      "Important":(party_imp_total + party_very_imp_total)/party_total, 
                      "Not Important":party_not_imp_total/party_total,
                           }}
    return pd.DataFrame.from_dict(totals)


```

```python
def getRespondentGroups(df):
    
    #Constituencies
    constituencies = mdf['Constituency'].dropna().unique().tolist()
    #States
    states = mdf['State'].dropna().unique().tolist()
    #Parties 
    parties = mdf['Party'].dropna().unique().tolist()
    #Genders
    genders = mdf['Respondent Gender'].dropna().unique().tolist()
    #Constituency Location Urban or Rural
    const_locations = mdf['MP Constituency Location'].dropna().unique().tolist()
    #Age Groups 
    age_groups = mdf['Age Group'].dropna().unique().tolist()
    #Religions
    religions = mdf['Religion'].dropna().unique().tolist()
    #Education Levels
    ed_levels = mdf['Education Level'].dropna().unique().tolist()
    #Castes
    castes = mdf['Caste'].dropna().unique().tolist()
    
    #Roof Types
    roof_types = mdf['Home Roof Type'].dropna().unique().tolist()
    
    #Wealth Index
    wlth_indices = mdf['Wealth Index'].dropna().unique().tolist()
    
    #Stove Types
    stove_types = mdf['Stove Type'].dropna().unique().tolist()
    
    #Two Wheeler Ownership 
    two_wheelers = mdf['Two-Wheeler Ownership'].dropna().unique().tolist()
    
    #Four Wheeler ownership
    four_wheelers = mdf['Four-Wheeler Ownership'].dropna().unique().tolist()
    
    #Live Stock Ownership
    livestock_owners = mdf['Live Stock Ownership'].dropna().unique().tolist()
    
    resp_groups = {"Constituency": constituencies, "State": states, "Party": parties, 
                   "Respondent Gender": genders, "MP Constituency Location": const_locations, 
                   "Age Group": age_groups, "Religion": religions, 
                   "Education Level": ed_levels, "Caste": castes, "Wealth Index":wlth_indices,
                   "Home Roof Type": roof_types, "Stove Type": stove_types,
                   "Two-Wheeler Ownership": two_wheelers, "Four-Wheeler Ownership":four_wheelers,
                   "Live Stock Ownership": livestock_owners}
    return resp_groups
```

```python
def getPolViews(df):
    
    #Candidates
    cands = mdf['The Candidate'].dropna().unique().tolist()
    #Candidates Party
    cand_parties = mdf["Candidate's Party"].dropna().unique().tolist()
    #Candidate's Caste
    cand_castes = mdf["Candidate's Caste/Religion"].dropna().unique().tolist()
    
    pol_views = {"The Candidate": cands, "Candidate's Party": cand_parties,
                "Candidate's Caste/Religion":cand_castes}
    
    return pol_views
```

```python
def getPercentage(cat, pol_issue,df):
    tot = len(df.index)
    subdf = df.loc[df[pol_issue] == cat]
    cat_tot = len (subdf.index)
    if tot > 0:
        return cat_tot / tot
    return 0
        
```

```python
def getMedian( pol_issue, df):
    #subdf = df[df[pol_issue].apply(lambda x: x.isnumeric())]
    subdf = df.copy()
    subdf = subdf.sort_values(by=[pol_issue])
    return subdf[pol_issue].median()
```

```python
def getDataFrame(key, vals,pviews, iviews):
    master_data = []
    
    for val in vals:
        
        if key == 'Party':
            subdf = mdf.loc[(mdf['Party'] == val) & (mdf["If Yes -  Your Candidate Won?"] == "Yes")]
        else:
            subdf = mdf.loc[(mdf[key] == val)]
                   
        for pview, values in pviews.items():
            for value in values:
                #pass
                pctage = getPercentage(value,pview,subdf)
                master_data.append([val,pview,value,pctage])
                #print(const," ", pview, " ", value, " ",pctage)
        
        for iview in iviews:
            #print(iview)
            median = getMedian(iview,subdf)
            #print(const, " ",iview, " ", "MEDIAN ",median)
            master_data.append([val,iview,"MEDIAN",median])
        
    return pd.DataFrame(master_data)
```

```python
def buildReportDataSets():
    groups = getRespondentGroups(mdf)
    pol_views = getPolViews(mdf)
    issue_views = getIssueViews(mdf)
    
    for key, value in groups.items():
        print("Building ",key," Data Set")
        filename = "./Data/"+key+"_Data.csv"  
        pdf = getDataFrame(key, value,pol_views, issue_views)
        pdf.to_csv(filename)
```
